```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
print(np.where(st['math score']>90))
```

```python

print(np.where(st['reading score']<25))
```

```python
print(np.where(st['writing score']<30))
```

```python
fig,ax=plt.subplots(figsize=(18,10))
ax.scatter(st['placement score'],st['placement offer count'])
plt.show()
```

```python
print(np.where((st['placement score']<50) & (st['placement offer count']>1)))
print(np.where((st['placement score']>85) & (st['placement offer count']<3)))
```

```python
st["gender"]=st["gender"].str.strip().map({"female":1,"male":0})
st
```

```python
Q1=st["reading score"].quantile(0.25)
Q3=st["reading score"].quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
outliers = st[(st["reading score"] < lower_bound)  |(st["reading score"] > upper_bound)]
print(outliers)
```
